```python
'''Problem Statement 7:
Customers arrive at a rate of 72 per hour to my shop. What is the probability of 𝑘
customers arriving in 4 minutes? a) 5 customers, b) not more than 3 customers, c) 
more than 3 customers. Give a pictorial representation of the same to validate your 
answer'''

'''A Poisson random variable “x” defines the number of successes in the experiment. 
This distribution occurs when there are events that do not occur as the outcomes of a definite number of outcomes.
Poisson distribution is used under certain conditions.'''

# the rate of customer arriving per hour = 72
# so, rate of customer arriving in 4 minute = 72/60 * 4 = 4.8 approx.
# so my lambda value is 4.8 for poisson probability distribution having formula = e** (-lambda) * lambda
from scipy.stats import poisson
print(poisson.pmf(5,4.8)) # finding probability for 5 customers in 4 minutes, pmf means probability mass function

print(poisson.cdf(5,3)) # finding probability for atleast 3 customers (not more than 3 customers) in 4 minutes

print(1 - poisson.cdf(5,3)) # for probability for more than 3 customers in 4 minutes
```
